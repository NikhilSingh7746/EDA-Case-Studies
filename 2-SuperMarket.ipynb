# Bivariate / Multivariate Analysis:


**1. Does the Cost of Goods Sold Affect the Ratings that the Customers Provide ? (Numerical-Numerical)**

```python
sns.scatterplot(x=data['Rating'] , y=data['cogs'] )
plt.show()  # No Relation
```

**2. Does Gross Income Affect the Ratings that the customers provide? (Numerical-Numerical)**

```python
sns.scatterplot(x=data['Rating'] , y=data['gross income'] )
plt.show()  # No Relation
```

**3. Find the Most profitable Branch as per Gross Income,(Numerical=categorical)**

```python
sns.barplot(x=data['Branch'] , y=data['gross income'] , palette='plasma' )
plt.show()
```

**4. Is there is any Relationship between Gender and gross Income (num-cat)**

```python
sns.barplot(data=data , x='Gender' , y='gross income' , palette='plasma')
plt.show()
```

```python
sns.boxplot(x=data['Gender'] , y=data['gross income']  , palette='plasma')
plt.show()
```

```python
# Multivariate
sns.boxplot(x=data['Gender'] , y=data['gross income'] , hue=data['Customer type'] , palette='plasma')
plt.show()

# Member customers contribute than normal Customers
```

```python
sns.barplot(x=data['Gender'] , y=data['gross income'] , hue=data['Customer type'] , palette='plasma')
plt.show()
```

5. Find the product line that Generates the most income (N-C)

```python
sns.barplot(x=data['Product line'] , y=data['gross income'] , palette='plasma')
plt.xticks(rotation=70)
plt.show()
```

6. Find the Highest Unit price In the Product Line (N-C)

```python
data.groupby('Product line')['Unit price'].max().sort_values(ascending=False)  # Highest Unit price
```

```python
sns.barplot(x=data['Product line'] , y=data['Unit price'] , palette='plasma')
plt.xticks(rotation=70)
plt.show()
```

**7. Find Different Payment Methods used by Customers Citywise.(C-C)**

```python
pd.crosstab(data['City'] , data['Payment'])
```

```python
# Stacked Bar Plot
pd.crosstab(data['City'] , data['Payment']).plot(kind='bar' , stacked=True)
plt.show()
```

```python
# HeatMap
sns.heatmap(pd.crosstab(data['City'] , data['Payment']) , annot=True)
plt.show()
```

8. Which Product Line is purchased in the Highest Quantity

```python
data.groupby('Product line')['Quantity'].sum().plot(kind='bar')
plt.show()
```

9. Display Daily sales by Day of the Weeks.

```python
data['days'] = data['Date'].dt.day_name()

data.groupby('days')['Total'].sum().sort_values(ascending=False)
```

```python
sns.barplot(x=data['days'] , y=data['Total'] , palette='plasma')
plt.show()
```

```python
# or
days = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

data['days_of_week'] = data['Date'].dt.dayofweek.map(days)

data.groupby('days_of_week')['Total'].sum().sort_values(ascending=False)
```
