19. Classify Movies Based on Ratings [Excellent, Good and Average]

```python
def Rating_classifier(x):
  if x['Rating'] >= 8:
    x['Rating_Category'] =  ' Excellent'
  elif x['Rating'] >=6 and x['Rating'] < 8:
    x['Rating_Category'] = 'Good'
  else:
    x['Rating_Category'] = 'Average'
  return x
```

```python
df.apply(Rating_classifier , axis=1)
```

21. Find Unique Values From Genre

```python
df['Genre'].nunique()
```

```python
genres = df['Genre'].unique()
genres
```

To Get Detailed Genres

```python
genres1 = [i for i in genres]
genres1
```

```python
genres2 = [i for i in genres1]
genres2

unique_genres_r = []
for i in genres2:
  for j in i.split(','):
    unique_genres_r.append(j)
```

```python
unique_genres = list(set(unique_genres_r))
unique_genres
```
