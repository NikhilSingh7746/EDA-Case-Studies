14. How Many People Have a Credit Card That Expires in 2020 ?

```python
# using datetime
df['CC Exp Date'] = pd.to_datetime(df['CC Exp Date'] , format='%m/%y')
df[df['CC Exp Date'].dt.year==2020]['CC Exp Date'].count()
```

```python
# or
df['CC Exp Date'].apply(lambda x : x.split('/')[1] == '20').sum()
```

```python
# or
def fun():
  count = 0
  for date in df['CC Exp Date']:
    if date.split('/')[1] == '20':
      count+=1
  print(count)

fun()
```
